```python
import numpy as np
import scipy.misc
from scipy.fftpack import dct, idct
import sys
from PIL import Image
import matplotlib.pyplot as plt
```

```python
bmp = Image.open("LENNA.bmp")
plt.imshow(bmp,"gray")
```

```python
# 2次元FFT
lenna_F = np.fft.fft2(bmp)
```

```python
# 象限のシフトなし
plt.imshow(np.log10(np.abs(lenna_F)))
```

```python
# 象限のシフトあり
# 配列の第1象限と第4象限、第2象限と第3象限をそれぞれ入れ替える
lenna_F = np.fft.fftshift(lenna_F)
plt.imshow(np.log10(np.abs(lenna_F)))
```
